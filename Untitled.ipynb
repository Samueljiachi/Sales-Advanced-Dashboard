```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
df["Day"] = df["Order Date"].dt.day

```

```python
df.head()
```

```python
cols = list(df.columns)
```

```python
cols
```

```python
cols.insert(cols.index("Order Date") +1,cols.pop(cols.index("Year")))
cols.insert(cols.index("Order Date") +2,cols.pop(cols.index("Month")))
cols.insert(cols.index("Order Date") +3,cols.pop(cols.index("Day")))
```

```python
df = df[cols]
```
